# Data Cleaning

1.delete duplicated Lines

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

2.detection Values NaN

```python
df.isnull().sum()
```

3.Delete useless Columns

```python
df.columns
```

```python
deleted_columns = ['id', 'imdb_id', 'budget', 'revenue',
       'cast', 'homepage', 'director', 'tagline', 'keywords', 'overview',
       'runtime', 'production_companies', 'release_date']
df.drop(deleted_columns, axis=1 ,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

4.Recoding of Column "genres"

```python
list_genre = []
for i in df['genres']:
    i = i.split('|')
    list_genre.append(i[0])
list_genre
```

```python
df['genres_adj'] = list_genre
df
```

```python
df.drop("genres", axis=1, inplace=True)
```

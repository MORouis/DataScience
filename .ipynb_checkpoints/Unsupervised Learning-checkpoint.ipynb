# 1. K-Means Clustering

cherche la position des centres qui minimise la distance entre les points d'un cluster et le centre de ce dernier (minimiser la variance des clusters)

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Génération de données
X, y = make_blobs(n_samples = 100, centers = 3, cluster_std=0.4, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
model.predict(X)
plt.scatter(X[:,0], X[:,1], c=model.predict(X))
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c='r') #cluster_centers_: position des centroids(3 centroides + 2 variables)
model.inertia_ #la somme des distances entre le pts_cluster et le centroids(Cout)
```

```python
inertia = []
K_range = range(1, 20)
for K in K_range:
    model = KMeans(n_clusters=K).fit(X)
    inertia.append(model.inertia_)
```

```python
plt.plot(K_range, inertia)
plt.xlabel("nombre de clusters")
plt.ylabel("Cout du modele (Inertia)")
```

# 2. Isolation Forest Anomaly Detection

consiste à détecter dans notre dataset les échantillons dans les caractéristiques X sont trés éloigné de celle des autres échantillons

```python
 X, y = make_blobs(n_samples=50, centers=1, cluster_std=0.1, random_state=0)
X[-1,:] = np.array([2.25, 5])

plt.scatter(X[:,0], X[:, 1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
model = IsolationForest(contamination=0.01) #1% de déchets dans la dataset doit etre identifiée
model.fit(X)
plt.scatter(X[:,0], X[:, 1], c=model.predict(X))
```

# Application décontamination Digits

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
images = digits.images
X = digits.data
y = digits.target

print(X.shape)
```

```python
#plt.imshow(images[58])
model = IsolationForest(random_state=0, contamination=0.02)
```

```python
model.fit(X)
```

```python
outliers = model.predict(X) == -1 #(-1): data représente d'anomalie; (1): normale
outliers
```

```python
plt.figure(figsize=(12, 3))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(images[outliers][i])
    plt.title(y[outliers][i])
```

# 3. PCA Réduction de Dimension

Le principe est de réduire la complexité superflue d'un dataset en projetant ses données dans un espace de plus petite dimension (un espace avec moins de variable)

==> Acclérer l'apprentissage de la machine

==> éviter Overfitting

# 3.1/ visualisation de données 

```python
from sklearn.decomposition import PCA
```

```python
X.shape
```

```python
model = PCA(n_components=2)
#model.fit_transform(X).shape
X_reduced = model.fit_transform(X)
```

```python
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=y)
plt.colorbar() #0 & 1 so far(different shape)
```

```python
model.components_.shape #deux composantes(chaque composant contient 64 valeurs )
```

```python
plt.figure()
plt.xlim(-30, 30)
plt.ylim(-30, 30)

for i in range(100):
    plt.text(X_reduced[i,0], X_reduced[i,1], str(y[i]))
```

# 3.2/ Compression de données 

==> accélérer l'apprentissage de la machine

réduire le nombre de variable en conservant 95%-99% de la variance

```python
model = PCA(n_components=64)
X_reduced = model.fit_transform(X)
```

```python
plt.plot(np.cumsum(model.explained_variance_ratio_)) #examiner quel est le pourcentage de variance preservé par chacune de nos composantes; cumsum(rendre resultat en pourcentage )
np.argmax(np.cumsum(model.explained_variance_ratio_) > 0.99)
```

```python
model = PCA(n_components=40) #plus on rajoute de composantes, plus la qualité de l'image est préservée
X_reduced = model.fit_transform(X)
```

```python
X_recovered = model.inverse_transform(X_reduced) #observer image compressé
```

```python
plt.imshow(X_recovered[0].reshape((8,8)))
```

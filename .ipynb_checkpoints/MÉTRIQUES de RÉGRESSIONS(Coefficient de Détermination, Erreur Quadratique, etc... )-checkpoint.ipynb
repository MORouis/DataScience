```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

# Quand utiliser la MSE plutôt que la MAE:

MSE→vous accordez une grande importance aux grandes erreurs.

MAE→l'importance d'une erreur est linéaire avec son amplitude.Si le Dataset contient des valeurs
aberrantes(outliers).

MAE(selon tri) = median{|y_vrai- y_pred|} → performant en cas de détection d Outliers

```python
y=np.array([1])
y_pred=np.array([1])

print('MAE',mean_absolute_error(y,y_pred))
print('MSE',mean_squared_error(y,y_pred))
```

```python
y=np.array([1, 2])
y_pred=np.array([5, 2])

print('MAE',mean_absolute_error(y,y_pred))
print('MSE',mean_squared_error(y,y_pred))
print('RMSE',np.sqrt(mean_squared_error(y,y_pred)))
```

```python
y=np.array([1, 2, 2, 3, 5, 2])
y_pred=np.array([5, 2, 2, 5, 7, 1000])

print('MAE',mean_absolute_error(y,y_pred))
print('MSE',mean_squared_error(y,y_pred))
print('median abs error',(median_absolute_error(y,y_pred)))
```

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

boston=load_boston()
X=boston.data
y=boston.target

model=LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

plt.scatter(X[:,5],y,label='y')
plt.scatter(X[:,5],y_pred,alpha=0.8,label='y_pred')
plt.legend()
```

Distribution exponentielle

```python
err_hist = np.abs(y - y_pred)
plt.hist(err_hist, bins=50)
plt.show()
```

```python
err_hist2 = (y - y_pred)**2
plt.hist(err_hist2, bins=50)
plt.show()
```

coefficient de détermination R2
R2 = 1 - (Σ(Yvrai-Ypred)²/Σ(vrai-(Y^vrai)²) ; Y^vrai:moyenne de yvrai
==> évalue la performance du modèle par rapport au niveau de variation présent dans les données.

```python
model.score(X,y) #ici, le modèle décrit 74% des variations du prix de l'immobilier
```

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

boston=load_boston()
X=boston.data
y=boston.target

model=LinearRegression()
model.fit(X,y)
model.score(X,y)
                          
                    
cross_val_score(LinearRegression(),X,y,cv=3,
                scoring='neg_mean_absolute_error')
```
